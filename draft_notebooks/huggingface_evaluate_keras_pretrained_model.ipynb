```python
import numpy as np
from src.data.read_data import SAMPLE_IMAGE, SAMPLE_LABEL
```

```python
image = np.expand_dims(SAMPLE_IMAGE, axis=0)
```

```python
predictions = model.predict(image)
```

```python
import evaluate
import json
from PIL import ImageColor
from src.preprocessors.label import label_to_categorical
```

```python
with open("./data/class_list.json") as f:
    class_colors_mapping_hex = json.load(f)
class_colors_mapping_rgb = {}
for color, label in class_colors_mapping_hex.items():
    class_colors_mapping_rgb[label] = ImageColor.getcolor(color, "RGB")
cat = label_to_categorical(SAMPLE_LABEL, class_colors_mapping_rgb)
```

```python
model.evaluate(x=image, y=cat)
```

```python
acc = evaluate.load("accuracy")
acc.compute(predictions = predictions, references = cat)
```
